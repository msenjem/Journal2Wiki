```python
articles_cleaned = "/gdrive/My Drive/CSCI 5541 Group/Project/articles_cleaned.csv"
original = pd.read_csv(articles_cleaned)
original
```

```python
rouge = evaluate.load('rouge')

rouge_score = []

fname = "/gdrive/My Drive/CSCI 5541 Group/Project/pickles/wiki_pages_text_2-330.pkl"
with open(fname, 'rb') as handle:
    wiki_pages_text = pickle.load(handle)
    handle.close()

index = 1
for i in wiki_pages_text:
  index = index + 1
  r_temp = rouge.compute(predictions= ["".join(i)], references=[original["Wiki"][index]])
  rouge_score.append(r_temp)

df = pd.DataFrame(rouge_score)
excel_file = "rouge_scores.xlsx"
df.to_excel(excel_file, index=False)
```
